```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from datetime import datetime
from pandas.plotting import autocorrelation_plot

db = pd.read_csv("/content/drive/My Drive/datawebscrap/vgsales.csv", encoding= 'unicode_escape')

#Filtering
db['Year_datetime'] = pd.to_datetime(db['Year'], format='%Y')
db['Year2'] = db['Year_datetime'].dt.year

playstation = ['PS3','PS2', 'PSP']
select = db ['Platform'].isin(playstation)

db_playstation = db[select]
playstation_sales = db_playstation.groupby('Year2')
df_ps_sales = db_playstation.groupby('Year2')['Global_Sales'].sum().reset_index()

#Visualisasi
sns.set(style="whitegrid")
sns.set_palette('Accent')
plt.figure(figsize=(10, 6))
ax = sns.lineplot(x='Year2', y = 'Global_Sales', data = df_ps_sales)
ax.set_title('Penjualan Playstation ', loc = 'left', fontsize = 14)
ax.set_xlabel('Tahun', fontsize = 10)
ax.set_ylabel('Penjualan Dunia($)',fontsize = 10)
ax = ax
```

```python
#coba geeksforgeeks
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller

df = pd.read_csv("/content/drive/My Drive/datawebscrap/stock_data.csv",
                 parse_dates=True,
                 index_col="Date")

#Filtering
df.drop(columns='Unnamed: 0', inplace =True)

#Visualisasi

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='High', label='High Price', color='blue')

plt.xlabel('Date')
plt.ylabel('High')
plt.title('Share Highest Price Over Time')

plt.show()
```
